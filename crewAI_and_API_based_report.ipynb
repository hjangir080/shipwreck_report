```python
#code taken from openmateo.com
#marine forecast code and general weather forecast code both are being called and then merged together

import openmeteo_requests

import requests_cache
import pandas as pd
from retry_requests import retry

# Setup the Open-Meteo API clients with cache and retry on error
cache_session_forecast = requests_cache.CachedSession('.cache_forecast', expire_after = 3600)
retry_session_forecast = retry(cache_session_forecast, retries = 5, backoff_factor = 0.2)
forecast_client = openmeteo_requests.Client(session = retry_session_forecast)

cache_session_marine = requests_cache.CachedSession('.cache_marine', expire_after = 3600)
retry_session_marine = retry(cache_session_marine, retries = 5, backoff_factor = 0.2)
marine_client = openmeteo_requests.Client(session = retry_session_marine)

# User input for latitude and longitude (optional)
while True:
  try:
    latitude = float(input("Enter latitude (decimal degrees): "))
    longitude = float(input("Enter longitude (decimal degrees): "))
    # Validate input range for latitude and longitude (-90 to 90, -180 to 180)
    if latitude < -90 or latitude > 90 or longitude < -180 or longitude > 180:
      print("Invalid input. Latitude must be between -90 and 90, longitude between -180 and 180.")
    else:
      break
  except ValueError:
    print("Invalid input. Please enter a valid decimal number.")

# Forecast API call (land-based weather)
url_forecast = "https://api.open-meteo.com/v1/forecast"
params_forecast = {
  "latitude": latitude,
  "longitude": longitude,
  "hourly": ["temperature_2m", "wind_speed_10m"]
}
responses_forecast = forecast_client.weather_api(url_forecast, params=params_forecast)

# Process first location for forecast data
response_forecast = responses_forecast[0]
print(f"Coordinates {response_forecast.Latitude()}°N {response_forecast.Longitude()}°E")
print(f"Timezone {response_forecast.Timezone()} {response_forecast.TimezoneAbbreviation()}")
print(f"Timezone difference to GMT+0 {response_forecast.UtcOffsetSeconds()} s")

# Process hourly data for forecast
hourly_forecast = response_forecast.Hourly()
hourly_temperature_2m = hourly_forecast.Variables(0).ValuesAsNumpy()
hourly_wind_speed_10m = hourly_forecast.Variables(1).ValuesAsNumpy()

hourly_data_forecast = {"date": pd.date_range(
  start = pd.to_datetime(hourly_forecast.Time(), unit = "s", utc = True),
  end = pd.to_datetime(hourly_forecast.TimeEnd(), unit = "s", utc = True),
  freq = pd.Timedelta(seconds = hourly_forecast.Interval()),
  inclusive = "left"
)}
hourly_data_forecast["temperature_2m"] = hourly_temperature_2m
hourly_data_forecast["wind_speed_10m"] = hourly_wind_speed_10m

forecast_dataframe = pd.DataFrame(data = hourly_data_forecast)

# Marine API call (wave height)
url_marine = "https://marine-api.open-meteo.com/v1/marine"
params_marine = {
    "latitude": latitude,
    "longitude": longitude,
    "hourly": "wave_height"  #can add more variables like wave direction etc.
}
responses_marine = marine_client.weather_api(url_marine, params=params_marine)

# Process first location for marine data
response_marine = responses_marine[0]
# Removed elevation print statement (not provided by Marine API)

# Process hourly data for marine (wave height)
hourly_marine = response_marine.Hourly()
hourly_wave_height = hourly_marine.Variables(0).ValuesAsNumpy()

hourly_data_marine = {"date": pd.date_range(
  start = pd.to_datetime(hourly_marine.Time(), unit = "s", utc = True),
  end = pd.to_datetime(hourly_marine.TimeEnd(), unit = "s", utc = True),
  freq = pd.Timedelta(seconds = hourly_marine.Interval()),
  inclusive = "left"
)}
hourly_data_marine["wave_height"] = hourly_wave_height

marine_dataframe = pd.DataFrame(data = hourly_data_marine)

# Merge forecast and marine DataFrames (assuming 'date' is the common column)
merged_dataframe = forecast_dataframe.merge(marine_dataframe, on='date', how='left')

print(merged_dataframe.head(80)) #first 80 hours

```
