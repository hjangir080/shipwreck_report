```python
#code taken from openmateo.com
#marine forecast code and general weather forecast code both are being called and then merged together

import openmeteo_requests

import requests_cache
import pandas as pd
from retry_requests import retry

# Setup the Open-Meteo API clients with cache and retry on error
cache_session_forecast = requests_cache.CachedSession('.cache_forecast', expire_after = 3600)
retry_session_forecast = retry(cache_session_forecast, retries = 5, backoff_factor = 0.2)
forecast_client = openmeteo_requests.Client(session = retry_session_forecast)

cache_session_marine = requests_cache.CachedSession('.cache_marine', expire_after = 3600)
retry_session_marine = retry(cache_session_marine, retries = 5, backoff_factor = 0.2)
marine_client = openmeteo_requests.Client(session = retry_session_marine)

# User input for latitude and longitude (optional)
while True:
  try:
    latitude = float(input("Enter latitude (decimal degrees): "))
    longitude = float(input("Enter longitude (decimal degrees): "))
    # Validate input range for latitude and longitude (-90 to 90, -180 to 180)
    if latitude < -90 or latitude > 90 or longitude < -180 or longitude > 180:
      print("Invalid input. Latitude must be between -90 and 90, longitude between -180 and 180.")
    else:
      break
  except ValueError:
    print("Invalid input. Please enter a valid decimal number.")

# Forecast API call (land-based weather)
url_forecast = "https://api.open-meteo.com/v1/forecast"
params_forecast = {
  "latitude": latitude,
  "longitude": longitude,
  "hourly": ["temperature_2m", "wind_speed_10m"]
}
responses_forecast = forecast_client.weather_api(url_forecast, params=params_forecast)

# Process first location for forecast data
response_forecast = responses_forecast[0]
print(f"Coordinates {response_forecast.Latitude()}°N {response_forecast.Longitude()}°E")
print(f"Timezone {response_forecast.Timezone()} {response_forecast.TimezoneAbbreviation()}")
print(f"Timezone difference to GMT+0 {response_forecast.UtcOffsetSeconds()} s")

# Process hourly data for forecast
hourly_forecast = response_forecast.Hourly()
hourly_temperature_2m = hourly_forecast.Variables(0).ValuesAsNumpy()
hourly_wind_speed_10m = hourly_forecast.Variables(1).ValuesAsNumpy()

hourly_data_forecast = {"date": pd.date_range(
  start = pd.to_datetime(hourly_forecast.Time(), unit = "s", utc = True),
  end = pd.to_datetime(hourly_forecast.TimeEnd(), unit = "s", utc = True),
  freq = pd.Timedelta(seconds = hourly_forecast.Interval()),
  inclusive = "left"
)}
hourly_data_forecast["temperature_2m"] = hourly_temperature_2m
hourly_data_forecast["wind_speed_10m"] = hourly_wind_speed_10m

forecast_dataframe = pd.DataFrame(data = hourly_data_forecast)

# Marine API call (wave height)
url_marine = "https://marine-api.open-meteo.com/v1/marine"
params_marine = {
    "latitude": latitude,
    "longitude": longitude,
    "hourly": "wave_height"  #can add more variables like wave direction etc.
}
responses_marine = marine_client.weather_api(url_marine, params=params_marine)

# Process first location for marine data
response_marine = responses_marine[0]
# Removed elevation print statement (not provided by Marine API)

# Process hourly data for marine (wave height)
hourly_marine = response_marine.Hourly()
hourly_wave_height = hourly_marine.Variables(0).ValuesAsNumpy()

hourly_data_marine = {"date": pd.date_range(
  start = pd.to_datetime(hourly_marine.Time(), unit = "s", utc = True),
  end = pd.to_datetime(hourly_marine.TimeEnd(), unit = "s", utc = True),
  freq = pd.Timedelta(seconds = hourly_marine.Interval()),
  inclusive = "left"
)}
hourly_data_marine["wave_height"] = hourly_wave_height

marine_dataframe = pd.DataFrame(data = hourly_data_marine)

# Merge forecast and marine DataFrames (assuming 'date' is the common column)
merged_dataframe = forecast_dataframe.merge(marine_dataframe, on='date', how='left')

print(merged_dataframe.head(80)) #first 80 hours

```

```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
search_tool = SerperDevTool()
```

```python
from langchain_openai import ChatOpenAI
```

```python
researcher = Agent(
    role = "researcher",
    goal = "research and present",
    backstory = "People's lives depend on you, you need to help save them in whatever way possible. So you think like Sherlock Holmes.",
    verbose = True,
    tools = [search_tool],
    llm = ChatOpenAI(model_name="gpt-4o", temperature = 0.2)
)
```

```python
navigator = Agent(
    role = "navigator",
    goal = "navigate and find feasible paths",
    backstory = "You are a living Google Maps and can find all possible routes the ship could have taken from the last recorded position",
    verbose = True,
    tools = [search_tool],
    llm = ChatOpenAI(model_name="gpt-4o", temperature = 0.2)
)
```

```python
doctor = Agent(
    role = "doctor",
    goal = "Diagnose all possible conditions the survivors might be in.",
    backstory = "You are the best medical diagnostic and doctor and you can predict all possible illnesses or injuries the survivors might have.",
    verbose = True,
    tools = [search_tool],
    llm = ChatOpenAI(model_name="gpt-4o", temperature = 0.2)
)
```

```python
task_research = Task(
    description = "Conduct a comprehensive and very detailed research into what all might have happeneed to the ship based on old records of ships gone missing or being in accidents around the area (Coordinates 18.125°N 39.75°E).",
    expected_output = "5 such possible scenarios in detail along with past examples.",
    agent = researcher
)
```

```python
task_navigate = Task(
    description = "Find out all possible sea routes that the ship might have taken from the Coordinates 18.125°N 39.75°E. as well as look into which places near these coordinates are dangerous and accident prone.",
    expected_output = "all possible routes in detail along with all dangerous places along the way.",
    agent = navigator
)
```

```python
task_diagnose = Task(
    description = "Diagnose all possible conditions the survivors might be in and their treatments.",
    expected_output = "List and describe all possible illnesses or injuries the survivors might have and material and medicines required to treat them.",
    agent = doctor
)
```

```python
crew1 = Crew(
    agents = [researcher],
    tasks = [task_research],
    verbose = True
)
```

```python
crew2 = Crew(
    agents = [navigator],
    tasks = [task_navigate],
    verbose = True
)
```

```python
crew3 = Crew(
    agents = [doctor],
    tasks = [task_diagnose],
    verbose = True
)
```

```python
result = crew1.kickoff()

print("########################RESEARCH#################################")
print(result)
```

```python
route_result=crew2.kickoff()
print("########################ROUTES#################################")
print(route_result)
```

```python
med_result=crew3.kickoff()
print("######################MEDICAL REQUIREMENT###################################")
print(med_result)
```

```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4
from reportlab.lib.units import cm


# Create a canvas object with A4 page size
c = canvas.Canvas("my_report.pdf", pagesize=A4)

# Set font properties (adjust font name, size, and color as needed)
c.setFont("Helvetica", 12)
c.setFillColorRGB(0, 0, 0)  # Set text color to black

# Define the starting point for the text (equal left and right margins)
left_margin = 1 * cm
x, y = left_margin, 29.7 * cm - 2 * cm  # Start at 2 cm from the top of the page

# Split the text into lines and add them to the canvas
for line in final_report.splitlines():
    # Check if the remaining space on the page is insufficient for the next line
    if y < 2 * cm:
        c.showPage()  # Start a new page
        y = 29.7 * cm - 2 * cm  # Reset vertical position for the new page

    # Draw the line of text
    c.drawString(x, y, line.rstrip())  # Remove trailing spaces
    y -= 12  # Adjust vertical position for the next line (line height)

# Save the PDF document
c.save()
print("PDF created successfully: my_report.pdf")
```
